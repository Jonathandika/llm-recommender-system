```python
import os
import pandas as pd
```

## Book Dataset

```python
book_dfs = []

for i in os.listdir('../data'):
    if i.startswith('book') and i.endswith('csv'):
        book_dfs.append(pd.read_csv('../data/' + i))

book_df = pd.concat(book_dfs)
```

```python
book_df.set_index('Id', inplace=True)
```

```python
book_df.sort_index(inplace=True)
```

```python
book_df.info()
```

```python
book_df
```

```python
sample_books = book_df[book_df['Description'].notnull()].sample(100)[['Name', 'Description']]
sample_books
```

```python
sample_books.to_csv('../data/sample_books.csv')
```

```python
book_df.to_parquet('../data/book.parquet')
```

## User Rating Dataset

```python
user_rating_dfs = []

for i in os.listdir('../data'):
    if i.startswith('user'):
        user_rating_dfs.append(pd.read_csv('../data/' + i))

user_rating_df = pd.concat(user_rating_dfs)
```

```python
user_rating_df.info()
```

```python
user_rating_df
```

```python
# Join user_rating_df and book_df to get the book_id based on name
book_df_reset = book_df.reset_index().rename(columns={'Id': 'book_id'})
user_rating_complete = pd.merge(user_rating_df, book_df_reset, left_on='Name', right_on='Name', how='left')[['Name', 'book_id', 'Rating_x']]
user_rating_complete.rename(columns={'Rating_x': 'Rating'}, inplace=True)
```

```python
user_rating_complete['Rating'].unique()
```

```python
rating_to_score = {
    'it was amazing': 5,
    'really liked it': 4,
    'liked it': 3,
    'it was ok': 2,
    'did not like it': 1,
    "This user doesn't have any rating" : -1
}

user_rating_complete['Rating'] = user_rating_complete['Rating'].apply(lambda x: rating_to_score[x])
```
